# MaskedFaceRecognition
## Abstract
This research is trying to analyze masked faces of subjects, and recognize them based on unmasked faces. We will use known face datasets and using Mask Adding algorithm, In aiming to make a new algorithm that will approve the connection between masked faces to unmasked.

## The problem
Nowadays, face recognition is a popular way of authentication. It is widely used in our every day with ways of accessing our phone by showing our face to the camera to match against our predefined access list or even crossing the border to match our passport's photo.

The ongoing COVID-19 pandemic has led us all to use face masks on a daily basis as a way to prevent the spread of the disease, which is highly contagious and airborne. But unfortunately, this practice has broken the worldwide face recognition algorithms which use features from all around our faces to identify us and prevents us from using face recognition in public.

With that, the need for algorithms that can recognize faces with masks, mainly for public usage, has grown rapidly. Moreover, the algorithms should work with the variety of face masks available today.

## The solution
We want to create an arteficial intelligence model and let it learn from images of unmasked faces found in public datasets. The model will learn the pattern of faces from the dataset. The model will produce an identifier for each person in a photo by the features of its face, then we will be able to compare images and decide if the person is the same one. Of course, one of the images will have a person without a mask and the other with (Images without a mask should work as well).

Validation of a person with partial visibility can be useful in other cases other than a pandemic (or just high personal hygiene), like in cases where a person might need to be verified with work or style equipment (like sunglasses) or in cases where identification aginst a database is needed (like police in events of partially masked robbery or an object hiding part of the face because of the environment of the image)

## Face recognition and Facial features

Indeed, there are numerous factors that cause the appearance of the face to vary.
The sources of variation in the facial appearance can be categorized into two groups: intrinsic factors and extrinsic ones [36].  
<b>Intrinsic factors</b> are due purely to the physical nature of the face and are independent of the observer.
These factors can be further divided into two classes: intrapersonal and interpersonal [37].  
    <b>Intrapersonal factors</b> are responsible for varying the facial appearance of the same person, some examples being age, facial expression and facial paraphernalia (facial hair, glasses, cosmetics, etc.).  
    <b>Interpersonal</b> factors, however, are responsible for the differences in the facial appearance of different people, some examples being ethnicity and gender.  
<b>Extrinsic factors</b> cause the appearance of the face to alter via the interaction of light with the face and the observer.
These factors include illumination, pose, scale and imaging parameters (e.g., resolution, focus, imaging, noise, etc.).  
  
Evaluations of recognition techniques from past research, such as the FERET, FRVT, FAT and such, have confirmed that age variations, illumination variations and pose variations are three major problems plaguing current face recognition systems [42]. Other challanges include occlusion of the face, face expressions and low resolution inputs.

### Feature-based recognition

Facial feature detection in images and video deals with estimating the position of prominent features of a human face on a facial image. Feature-based approaches first process the input image to identify and extract (and measure) distinctive facial features, and then compute the geometric relationships among those facial points, thus reducing the input facial image to a vector of geometric features. Later, standard statistical pattern recognition techniques are then employed to match faces using these measurements.  
The detected features are, in most cases, the eyes, the eyebrows, the nose and the mouth. The result of the detection can be either a list of characteristic fiducial points on those features, namely the eye corners, the center of the iris, the corners or the center of the eyebrows, the nose tip, the nostrils, the corners or the center of the mouth, or the  corresponding facial image region, represented by its region of interest (ROI) described either by its bounding box or more  accurately by its contour (e.g. lips contour).

#### Traditional algorithms approach

The traditional approach is to use well-established CV techniques such as feature descriptors (SIFT, SURF, HOG, BRIEF, etc.) for object detection. Such features might come in handy also with the initial mission of detecting a face in an image, like the Viola-Jones algorithm (which has its own features for the task). Then, a step called feature extraction was carried out for tasks such as image
classification. Features are small “interesting”, descriptive or informative patches in images. Several CV algorithms, such as edge detection, corner detection or threshold segmentation may be involved in this step. As many features as practicable are
extracted from images and these features form a definition (known as a bag-of-words) of each object class. At the deployment stage, these definitions are searched for in other images. If a significant number of features from one bag-of-words are in another image, the image is classified as containing that specific object (i.e. chair, horse, etc.). [8]

With that being said, face recoginition is a well-studied field of computer vision and past research offers various techniques and goals for feature extraction, as mentioned before. By using traditional computer vision algorithms to extract landmarks from facial image and using traditional classifier algorithms, such as SVM, KNN, Naive Bayes, etc., we can use effective tools to solve this complex-looking problem.

#### DL approach

The development of CNNs has had a tremendous influence in the field of CV in recent years and is responsible for a big jump in the ability to recognize objects [9]. This burst in progress has been enabled by an increase in computing power, as well as an increase in the amount of data available for training neural networks. The recent explosion in and wide-spread adoption of various deep-neural network architectures for CV is apparent in the fact that the seminal paper ImageNet Classification with Deep Convolutional Neural Networks has been cited over 3000 times [8]

While holistic (complete end-to-end) face processing solutions exist, most published works treat each subproblem separately. It is possible to study and evaluate each challenge in isolation, however, a combination of these is typically applied as a pipeline for ultimately performing face recognition or facial expression analysis. 
Face detection, locating the faces in a given image, is essential for many tasks in face processing. Head pose estimation determine the head orientation relative to the camera, and is important to a variety of face processing approaches. Another
crucial task for many applications is face alignment, which determines the geometric structure of the face. It is directly
applied to 3D face reconstruction, the recovery of the shape and appearance of the face, which can assist face recognition,
as it is invariant to scale and robust to occlusion.[7]

#### Attempt 1 - dlib + cv2
We first tried to use dlibs' frontal face detector to find landmarks on a face and then using opencv to draw a mask on the face of the person by their detected landmarks.

Here's a snippet of code of how it's done:

```python
# Necessary imports
import cv2
import dlib
import numpy as np
import os
import imutils
from matplotlib import pyplot
import matplotlib.image as mpimg

path = "data\\Nicolas_Cage_unmasked1.jpg"
color_blue = (254, 207, 110)

# Loading the image and resizing, converting it to grayscale
img = cv2.imread(path)
img = imutils.resize(img, width=500)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Initialize dlib's face detector
detector = dlib.get_frontal_face_detector()
p = "data\\shape_predictor_68_face_landmarks.dat"
# Initialize dlib's shape predictor
predictor = dlib.shape_predictor(p)

# Detecting faces in the grayscale image and creating an object - faces to store the list of bounding rectangles coordinates
# The "1" in the second argument indicates that we should upsample the image 1 time.  
# This will make everything bigger and allow us to detect more faces
face = detector(gray, 1)[0]  # Currently the image has only one face
landmarks = predictor(gray, face)

# for n in range(0,68):
#     x = landmarks.part(n).x
#     y = landmarks.part(n).y
#     img_landmark = cv2.circle(img, (x, y), 4, (0, 0, 255), -1)

points = []
for i in range(1, 16):
    point = [landmarks.part(i).x, landmarks.part(i).y]
    points.append(point)
# print(points)

# Coordinates for the additional 3 points for wide, high coverage mask - in sequence
mask_a = [((landmarks.part(42).x), (landmarks.part(15).y)),
          ((landmarks.part(27).x), (landmarks.part(27).y)),
          ((landmarks.part(39).x), (landmarks.part(1).y))]

# Coordinates for the additional point for wide, medium coverage mask - in sequence
mask_c = [((landmarks.part(29).x), (landmarks.part(29).y))]

# Coordinates for the additional 5 points for wide, low coverage mask (lower nose points) - in sequence
mask_e = [((landmarks.part(35).x), (landmarks.part(35).y)),
          ((landmarks.part(34).x), (landmarks.part(34).y)),
          ((landmarks.part(33).x), (landmarks.part(33).y)),
          ((landmarks.part(32).x), (landmarks.part(32).y)),
          ((landmarks.part(31).x), (landmarks.part(31).y))]

fmask_a = np.array(points + mask_a, dtype=np.int32)
fmask_c = np.array(points + mask_c, dtype=np.int32)
fmask_e = np.array(points + mask_e, dtype=np.int32)

pyplot.subplot(2, 3, 2)
pyplot.imshow(mpimg.imread(path), interpolation='nearest')
pyplot.axis('off')
for index, mask_type in enumerate([fmask_a, fmask_c, fmask_e]):
    img2 = cv2.polylines(img.copy(), [mask_type], True, color_blue, thickness=2, lineType=cv2.LINE_8)
    # Using Python OpenCV – cv2.fillPoly() method to fill mask
    # change parameter [mask_type] and color_type for various combination
    img3 = cv2.fillPoly(img2, [mask_type], color_blue, lineType=cv2.LINE_AA)
    pyplot.subplot(2, 3, 4 + index)
    pyplot.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB), interpolation='nearest')
    pyplot.axis('off')
pyplot.show()
```

### Prepare data for algorithms
Now we need to make pairs images of same person and pairs of images of different persons in order to run our algorithms

First, we will pair as many masked and unmasked images of the same person as possible

```python
people=merged_dataset["name"].unique()
total_pairs = []
def get_indexes(name, mask=True):
    if mask:
        mask_filter = merged_dataset["is_masked"]==True
    else:
        mask_filter = merged_dataset["is_masked"]==False
    person_filter = merged_dataset["name"] == name
    return list(merged_dataset[person_filter & mask_filter]["path"])
def get_pairs():  
    all_pairs = {}
    for name in people:
        pairs = list(zip(get_indexes(name,mask=True), get_indexes(name, mask=False)))
        if pairs:
            all_pairs[name] = pairs
    return all_pairs


#example of usage:
all_pairs_by_person = get_pairs()
print("Number of people:", len(all_pairs_by_person))
print("Number of pairs:", sum([len(person_pairs) for person_pairs in all_pairs_by_person.values()]))
```

Now, we will take these valid pairs and split them into two groups, the first will stay as valid pairs and the other will be pair of images of different people (while the first image is masked and the second is unmasked)

```python
import random

def split_pairs(pairs_by_person):
    all_pairs = []
    for person, person_pairs in pairs_by_person.items():
        for pair in person_pairs:
            masked_path, unmasked_path = pair
            all_pairs.append((masked_path, unmasked_path, person))
    random.shuffle(all_pairs)
    
    pos_pairs = all_pairs[:len(all_pairs) // 2]
    neg_pairs = []
    rest = all_pairs[len(all_pairs) // 2:]
    while len(set([pair[2] for pair in rest])) > 1:
        i = 0
        next_rest = []
        while i < len(rest):
            if i+1 == len(rest):
                next_rest.append(rest[-1])
                break
            pair1 = rest[i]
            pair2 = rest[i+1]
            if pair1[2] != pair2[2]:
                neg_pairs.append((pair1[0], pair2[1]))
                neg_pairs.append((pair2[0], pair1[1]))
            else:
                next_rest.append(pair1)
                next_rest.append(pair2)
            i+=2
        rest = next_rest
    pos_pairs += rest
    pos_pairs = [(pair[0], pair[1]) for pair in pos_pairs]
    return pos_pairs, neg_pairs
                
pos_pairs, neg_pairs = split_pairs(all_pairs_by_person)
```

```python
def dlib_detector_faces(detector, pairs):
    masked_results = []
    unmasked_results = []
    executions = 0
    errors = 0
    execution_time = time.time()
    for person, person_pairs in pairs.items():
        for pair in person_pairs:
            try:
                masked_path, unmasked_path = pair

                img = cv2.imread(masked_path)
                gray = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY)
                faces = detector(gray)

                img2 = cv2.imread(unmasked_path)
                gray2 = cv2.cvtColor(src=img2, code=cv2.COLOR_BGR2GRAY)
                faces2 = detector(gray2)

                masked_results.append((len(faces) > 0))
                unmasked_results.append((len(faces2) > 0))
            except Exception as e:
                print(e)
                errors += 1
            else:
                executions += 2
                execution_time
    print("Average execution time:", (time.time() - execution_time) / executions)
    print("Errors count:", errors)
    return masked_results, unmasked_results

masked_results, unmasked_results = dlib_detector_faces(detector=dlib_hog_detector, pairs=all_pairs_by_person)
```

## OpenCV SIFT
OpenCV is a public image processing library. The python implementation is referred to as cv2 and contains variaty of computer vision and image processing algorithms.
SIFT ORB and BRISK are image matching algorithms based on histogram of gradients, these algorithms find keypoints in the images and construct a descriptor for the image.

Let's define functions that will help us execute the feature matchers.
First for the positive tests, then for the negative tests.

```python
def cv_image_matching_knn(feature, matcher, pairs):
    results = []

    for masked_path, unmasked_path in pairs:
        # read images
        try:
            img1 = cv2.resize(cv2.imread(str(masked_path)), (224,224))
            img2 = cv2.resize(cv2.imread(str(unmasked_path)), (224,224))

            img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
            img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

            keypoints_1, descriptors_1 = feature.detectAndCompute(img1,None)
            keypoints_2, descriptors_2 = feature.detectAndCompute(img2,None)
            if not descriptors_1.size or not descriptors_2.size:
                print("nope")
                continue
            k_matches = matcher.knnMatch(descriptors_1, descriptors_2, k=2)

            # Apply ratio test
            good = 0
            for m,n in k_matches:
                if m.distance < 0.75 * n.distance:
                    good+=1
            results.append([(masked_path, unmasked_path), good])
        except Exception as e:
            print(e)
    return results
```

Now let's define functions to visualize our efforts

```python
def plot_percent_hist(results, results_neg):
    data = [(x, "positive") for x in results]+[(x, "negative") for x in results_neg]
    total_results = pd.DataFrame(data=data, columns=["distance", "type"])
    sns.histplot(total_results, x='distance', hue='type', element='step', kde=True, stat='percent')
```

```python
def plot_feature_heatmap(results, results_neg):
    total_results = len(results) + len(results_neg)
    TP=results.count(True) / total_results
    FN=results.count(False) / total_results
    FP=results_neg.count(True) / total_results
    TN=results_neg.count(False) / total_results
    array = [[TP,FN],
             [FP,TN]]

    df_cm = pd.DataFrame(array, ['same', 'different'], ['verified', 'unverified'])
    sns.set(font_scale=1.4) # for label size
    sns.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size

    plt.show()
```

### SIFT
We will define the feature descriptor of SIFT and use Flann (fast Knn) to match the keypoints that SIFT found. We will define our Flann based on the recommended values for integration with SIFT.

Let's start off with a simple example to demonstrate the usage of SIFT

```python
masked_path, unmasked_path = pos_pairs[100]

img1 = cv2.resize(cv2.imread(str(masked_path)), (224,224))
img2 = cv2.resize(cv2.imread(str(unmasked_path)), (224,224))

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)


sift = cv2.xfeatures2d.SIFT_create()
keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)

bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = bf.match(descriptors_1,descriptors_2)

img3 = cv2.drawMatches(img1,keypoints_1,img2,keypoints_2,matches[:10],None,flags=2)
plt.imshow(img3)
plt.show()
```

## VGGFACE

The VGGFace refers to a series of models developed for face recognition and demonstrated on benchmark computer vision datasets by members of the Visual Geometry Group (VGG) at the University of Oxford.

The model was based on the VGG16 network from a paper publish in 2015.

First we will define the neural network:

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import ZeroPadding2D,Convolution2D,MaxPooling2D
from tensorflow.keras.layers import Dense,Dropout,Softmax,Flatten,Activation,BatchNormalization
from tensorflow.keras.preprocessing.image import load_img,img_to_array,array_to_img
from tensorflow.keras.applications.imagenet_utils import preprocess_input
import tensorflow.keras.backend as K
```

```python
# Define VGG_FACE_MODEL architecture
model = Sequential()
model.add(ZeroPadding2D((1,1),input_shape=(224,224, 3)))
model.add(Convolution2D(64, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
model.add(ZeroPadding2D((1,1)))	
model.add(Convolution2D(128, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
model.add(Convolution2D(4096, (7, 7), activation='relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(4096, (1, 1), activation='relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(2622, (1, 1)))
model.add(Flatten())
model.add(Activation('softmax'))
```

```python
# Load VGG Face model weights
model.load_weights(r'..\data\vgg_face\vgg_face_weights.h5')
```

```python
from matplotlib import pyplot
from PIL import Image
from numpy import asarray
from scipy.spatial.distance import cosine
import pandas as pd
import pathlib
```

```python
def prepare_file(filename):
    img=load_img(filename, target_size=(224,224))
    img=img_to_array(img)
    img=np.expand_dims(img,axis=0)
    img=preprocess_input(img)
    return img

def get_embeddings(filenames):
    samples = [prepare_file(f) for f in filenames]
    global model
    return [np.squeeze(K.eval(model(s))).tolist() for s in samples]

# determine if a candidate face is a match for a known face
def is_match(known_embedding, candidate_embedding, thresh=0.5):
    # calculate distance between embeddings
    score = cosine(known_embedding, candidate_embedding)
    if score <= thresh:
        return True, score
    else:
        return False, score
```

```python
from keras.models import model_from_json
from inception_resnet_v1 import *
inception_model = InceptionResNetV1()
```

We will define helper function for the model, we are using L2 distance (Euclidean distance) on normalized vectors, as suggested by the papers and analysis of results of the model:

```python
def l2_normalize(x):
    return x / np.sqrt(np.sum(np.multiply(x, x)))

def findEuclideanDistance(source_representation, test_representation):
    euclidean_distance = source_representation - test_representation
    euclidean_distance = np.sum(np.multiply(euclidean_distance, euclidean_distance))
    euclidean_distance = np.sqrt(euclidean_distance)
    return euclidean_distance

def preprocess_image(image_path):
    img = load_img(image_path, target_size=(160, 160))
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    return img
```

```python
threshold = 0.5
```

```python
inception_results = []

# positive test
for masked_path, unmasked_path in pos_pairs:
    try:
        features1 = l2_normalize(inception_model.predict(preprocess_image(masked_path)))
        features2 = l2_normalize(inception_model.predict(preprocess_image(unmasked_path)))

        euclidean_distance = findEuclideanDistance(features1, features2)
        matched = euclidean_distance < threshold
        if matched:
            print("verified... they are same person")
        else:
            print("unverified! they are not same person!")
        inception_results.append([(masked_path, unmasked_path), matched, euclidean_distance])
        
    except Exception as e:
        print(e)
        continue
```

Without further a do we will start off with imports:

```python
from matplotlib import pyplot
from PIL import Image
from numpy import asarray
from scipy.spatial.distance import cosine
from keras_vggface.vggface import VGGFace
from keras_vggface.utils import preprocess_input
import pandas as pd
```

Next, we will define helper functions:

```python
# face verification with the VGGFace2 model
from matplotlib import pyplot
from PIL import Image
from numpy import asarray
from scipy.spatial.distance import cosine
# from mtcnn.mtcnn import MTCNN
from keras_vggface.vggface import VGGFace
from keras_vggface.utils import preprocess_input
import pandas as pd

def extract_face(filename, required_size=(224, 224)):
    image = Image.open(filename)
    image = image.resize(required_size)
    face_array = asarray(image)
    return face_array

def prepare_file(filenames):
    # extract faces
    faces = [extract_face(f) for f in filenames]
    # convert into an array of samples
    samples = asarray(faces, 'float32')
    # prepare the face for the model, e.g. center pixels
    return preprocess_input(samples, version=2)

vggface2_model = VGGFace(model='resnet50', include_top=False, input_shape=(224, 224, 3), pooling='avg')
def get_embeddings(filenames):
    # extract faces
    samples = prepare_file(filenames)
    # create a vggface model
    global vggface2_model
    # perform prediction
    yhat = vggface2_model.predict(samples)
    return yhat

# determine if a candidate face is a match for a known face
def is_match(known_embedding, candidate_embedding, thresh=0.5):
    score = cosine(known_embedding, candidate_embedding)
    if score <= thresh:
        print('>face is a Match (%.3f <= %.3f)' % (score, thresh))
        return True, score
    else:
        print('>face is NOT a Match (%.3f > %.3f)' % (score, thresh))
        return False, score
```

Now we are ready for the models' inference. We are using a cosine distance function as suggested.
